```python
import csv
total=0
ud=0
with open("datos-agricultura.csv", "r"   ,encoding="utf-8") as f:
    reader = csv.reader(f,delimiter=";")
    for row in reader:
        # Imprimimos los datos de la fila
        try:
            print(row)
            total=total+int(row[3])
            ud=ud+1
        except:
            print("")
            
print("el total es",total)

print("la media es ",(total/ud))
```

```python
import csv
total=0
numero_lineas=0
with open("datos-alumno.csv", "r",encoding="utf-8") as f:
    reader = csv.reader(f, delimiter=";")
    for row in reader:
        # Imprimimos los datos de la fila
        print(row[0], row[2], row[4], row[6])
        try:
            total=total+int(row[11])
            numero_lineas=numero_lineas+1
        except:
            print("")

        
print("la media es",(total/numero_lineas))
```
